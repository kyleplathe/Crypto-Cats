# Crypto Cat(egories) finding the balance.

In this assessment we will evaluate the performance among various crypto categories and compare the categories & variations to Bitcoin.

```python
from pycoingecko import CoinGeckoAPI   #Import CoinGeckoAPI
cg = CoinGeckoAPI()                    #Create the client 'cg'
from datetime import datetime          #Import the datetime module
from datetime import date
import time
import pandas as pd                    #Import pandas
from pathlib import Path
import numpy as np
import seaborn as sns
```

# Get Data

```python
coins = []
sdt = []
ndt = []
coin_portfolio_df = pd.DataFrame()

def get_data():
    coinlist()
    start_date()
    end_date()
    
    for coin in coins:
        crypto(coin, currency = 'usd')
        
    display(coin_portfolio_df)
    
        
        
def coinlist():               #Asks the user about the coins they want to get data from.
    global coins              #Make this global so that the list gets updated.
    n = int(input("How many coins do you want to retrieve data for? "))                                      #asks user about the number of coins for which data will be retrieved.
    coins = list(map(str,input("\nEnter the coin ids separated by a space : ").strip().split()))[:n]         #asks the user which coins to retrieve data for
    print(coins)
    return coins

def start_date():              #Asks the user for the earliest date on which to retreive data.
    global sdt
    user_in = input("On what date would you like the data to begin from? (yyyy-mm-dd): ")
    sdt = date.fromisoformat(user_in)
    return sdt

def end_date():                 #Asks the user for the latest data for which to retrieve 
    global ndt
    user_in = input("On what date would you like the data to end on? (yyyy-mm-dd): ")
    ndt = date.fromisoformat(user_in)
    return ndt
    
def crypto(coin, currency = 'usd'):             #Gets the crypto data and makes .csv files out of them. coin_id: from coins list, currency: default is 'usd'
    global coin_portfolio_df
    fromdate = sdt                                  #fromdate to unix_timestamp
    unix_fromdate = str(int(time.mktime(fromdate.timetuple())))
   
    todate = ndt                                     #todate to unix_timestamp
    unix_todate = str(int(time.mktime(todate.timetuple())))
    
    price_hist = cg.get_coin_market_chart_range_by_id(id = str(coin),          #gets data from coingecko
                                         vs_currency = str(currency),
                                         from_timestamp= unix_fromdate,
                                         to_timestamp= unix_todate)
    
    coin_df = pd.json_normalize(price_hist, record_path =['prices'])
    
    coin_df = coin_df.rename(columns = {0: 'date', 1: coin})
    
    coin_df['date'] = pd.to_datetime(coin_df['date'], unit='ms').dt.date
    
    coin_df.set_index("date", inplace=True)
    
    if coin == coins[0]:
        coin_portfolio_df = coin_df
    else:
        coin_portfolio_df = coin_portfolio_df.join(coin_df, how= 'inner')
        return coin_portfolio_df
   
    return coin_portfolio_df
    

```

# Stable Coin Portfolio

```python
# Calculate daily returns
daily_returns_stable_df = (stable_coin_df - stable_coin_df.shift(1)) / stable_coin_df.shift(1).dropna()

daily_returns_stable_df = daily_returns_stable_df.dropna()

weights = [1/10,1/10,1/10,1/10,1/10,1/10,1/10,1/10,1/10,1/10]
    
stable_returns = daily_returns_stable_df.dot(weights)

stable_returns.head()
```

# NFT Portfolio

```python
# Calculate daily returns
daily_returns_NFT_df = (NFT_df - NFT_df.shift(1)) / NFT_df.shift(1)

daily_returns_NFT_df = daily_returns_NFT_df.dropna()

weights = [1/10,1/10,1/10,1/10,1/10,1/10,1/10,1/10,1/10,1/10]
    
NFT_returns = daily_returns_NFT_df.dot(weights)

NFT_returns.head()
```

# DeFi Portfolio

```python
# Calculate daily returns
daily_returns_defi_df = (defi_df - defi_df.shift(1)) / defi_df.shift(1).dropna()

daily_returns_defi_df = daily_returns_defi_df.dropna()

weights = [1/10,1/10,1/10,1/10,1/10,1/10,1/10,1/10,1/10,1/10]
    
defi_returns = daily_returns_defi_df.dot(weights)

defi_returns.head()
```

# Metaverse Portfolio

```python
# Calculate daily returns
daily_returns_meta_df = (metaverse_df - metaverse_df.shift(1)) / metaverse_df.shift(1)

daily_returns_meta_df = daily_returns_meta_df.dropna()

weights = [1/10,1/10,1/10,1/10,1/10,1/10,1/10,1/10,1/10,1/10]
    
meta_returns = daily_returns_meta_df.dot(weights)

meta_returns.head()
```

# Bitcoin

```python
btc_df

# Calculate daily returns
daily_returns_btc_df = (btc_df - btc_df.shift(1)) / btc_df.shift(1)

daily_returns_btc_df = daily_returns_btc_df.dropna()

weights = [1]
    
btc_returns = daily_returns_btc_df.dot(weights)

btc_returns.head()
```

# Combine All Portfolios Returns

```python
combined_df = pd.concat([stable_returns, NFT_returns, defi_returns, meta_returns, btc_returns], axis="columns", join="inner")
combined_df = combined_df.rename(columns={0: "Stable Coins", 1: "NFTs", 2: "DeFi", 3: "Metaverse", 4: "Bitcoin"})

combined_df.head()
```

# Conduct Quantitative Analysis

In this section we calculate and visualize performance and risk metrics for the portfolios.

## Performance Analysis

### Plot the daily returns.

```python
# Plot daily returns of all portfolios

combined_df.plot(figsize=(10,5),title="Daily Return All Categories")
```

#### Calculate and plot cumulative returns.

```python
# Calculate cumulative returns of all portfolios
combined_cumulative_returns = (1 + combined_df).cumprod()

# Plot cumulative returns
combined_cumulative_returns.plot(figsize=(10,5), title="Cumulative Returns for all Categories")
```

### ------------------------------------------------

# Rick Analysis

Determine the risk of each portfolio:

```python
# Box plot to visually show risk
combined_df.plot.box(figsize=(25,5))
```

### Calculate Standard Deviations

```python
# Calculate the daily standard deviations of all portfolios
combined_daily_std = combined_df.std()
combined_daily_std
```

### Determine which portfolios are riskier than Bitcoin

```python
# Calculate  the daily standard deviation of Bitcoin
btc_daily_std = btc_df.std()
print(btc_daily_std)
print("\n")

#Sort highest risk portfolios descending
combined_daily_std_risk = combined_daily_std.sort_values(ascending=False)
combined_daily_std_risk
```

## Calculate the Annualized Standard Deviation

```python
# Calculate the annualized standard deviation (365 trading days)
combined_annualized_std = combined_daily_std * np.sqrt(365)
combined_annualized_std
```

## Rolling Statistics

### Calculate and plot rolling std for all portfolios with 21-day window

```python
# Calculate the rolling standard deviation for all portfolios using a 21-day window
rolling_combined = combined_df.rolling(window=21).std()

# Plot the rolling standard deviation
rolling_combined.plot(rot=45,figsize=(10,5), title="21-Day Rolling Standard Deviation for all Categories")
```

### Calculate and plot the correlation

```python
# Calculate the correlation
correlation = combined_df.corr()

# Display de correlation matrix

sns.heatmap(correlation, vmin=-1, vmax=1)
print(correlation)
```

### Calculate and Plot Beta for a chosen portfolio and Bitcoin

```python
# Calculate covariance of a single portfolio
covariance = combined_df['Metaverse'].cov(combined_df['Bitcoin'])
print(covariance)

# Calculate variance of Bitcoin
variance = combined_df['Bitcoin'].var()
print(variance)

# Computing beta
meta_beta = covariance / variance
print(meta_beta)

# Plot beta trend
rolling_covariance = combined_df['Metaverse'].rolling(window=21).cov(combined_df['Bitcoin'])
rolling_covariance.plot(figsize=(20, 10), title='Rolling 21-Day Covariance of Metaverse Returns vs. Bitcoin Returns')
```

# Sharpe Ratios

### Using the daily returns, calculate and visualize the Sharpe Ratios using a bar plot

```python
# Annualized Sharpe Ratios
sharpe_ratio = combined_df.mean() / combined_df.std()
annualized_sharpe_ratio = np.sqrt(365) * sharpe_ratio
annualized_sharpe_ratio
```

```python
# Visualize the sharpe ratios as a bar plot
annualized_sharpe_ratio.plot(kind="bar", title="Annualized Sharpe Ratios")
```
